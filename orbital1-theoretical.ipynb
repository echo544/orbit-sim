```python
# Import necessary libraries for plotting and numerical operations
import matplotlib.pyplot as plt
import numpy as np

# Create a figure and add a 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

# Star at origin
ax.scatter([0], [0], [0], color="y", s=100) # yellow point, size 100

# Plot planet's orbit
t = np.linspace(0, 2*np.pi, 100) # angle of planet in orbit, to be converted into x-y coords
x = np.cos(t)
y = np.sin(t)
ax.plot(x, y, zs=0, zdir='z', label="planet")

# Plot moon's orbit (orbits planet so is helical around star)
# we add the moon's position relative to the planet to the planet's position relative to the star to get the moon's position relative to the star
t2 = np.linspace(0, 2*np.pi, 1000)
x2 = 0.1 * np.cos(20*t2) + np.cos(t2) # make moon's orbit smaller than planet's and 20 times faster
y2 = 0.1 * np.sin(20*t2) + np.sin(t2)
ax.plot(x2, y2, zs=0, zdir="z", label="moon")

ax.legend()
plt.show()
```
